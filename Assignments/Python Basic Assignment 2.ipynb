## Question 3

```python
# list of each Boolean operators truth tables

# Here we will be taking 2 inputs as 0 or 1

l_inputs = [[0,0],[0,1],[1,0], [1,1]]
```

```python
print("AND OPERATIONS")

l_and = []

for i in l_inputs:
    l_operation = []
    input_1 = i[0]
    input_2 = i[1]
    l_operation.append(input_1)
    l_operation.append(input_2)
    l_operation.append(input_1 and input_2)
    
    l_and.append(l_operation)

print(l_and)
    
```

```python
print("OR OPERATIONS")

l_or = []

for i in l_inputs:
    l_operation = []
    input_1 = i[0]
    input_2 = i[1]
    l_operation.append(input_1)
    l_operation.append(input_2)
    l_operation.append(input_1 or input_2)
    
    l_or.append(l_operation)

print(l_or)
```

```python
print("NOT Operation")

l_not= []
l_input = [0,1]

for i in l_input:
    l_operation = []
    l_operation.append(i)
    l_operation.append((int(not i)))
    
    l_not.append(l_operation)

print(l_not)
```
